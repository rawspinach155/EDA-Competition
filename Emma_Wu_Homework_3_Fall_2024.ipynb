Remove Missing Values ([Tutorial](https://machinelearningmastery.com/handle-missing-data-python/), skip to Section 5)
*    Consider dropping rows with missing values if you have *enough* data (retaining at least 70% of data and having at least 100 observations).
*    Impute missing values with the `fillna` **method** with reasonable values.

```python
# Find the number of missing values per column
missing_values_count = df.isnull().sum()

# percentage of missing values per column
missing_values_percentage = (missing_values_count / len(df)) * 100

# Print the results
print("Number of missing values per column:\n", missing_values_count)
print("\nPercentage of missing values per column:\n", missing_values_percentage)
```

```python
# drop the columns with more than 40% missing values
threshold = 0.40 * len(df)
df = df.dropna(thresh=threshold, axis=1)

# drop the rows with more than 50% missing values
threshold = 0.50 * len(df.columns)
df = df.dropna(thresh=threshold, axis=0)

# print the new df and its updated rows/columns
df.head(10)

missing_values_count = df.isnull().sum()
missing_values_percentage = (missing_values_count / len(df)) * 100
print("Number of missing values per column:\n", missing_values_count)
print("\nPercentage of missing values per column:\n", missing_values_percentage)
```

```python
# drop rows with missing values
if len(df) * 0.7 >= 100:
    df.dropna(inplace=True)
    print("Rows with missing values dropped.")
print(df.isnull().sum())
```

# Data Exploration

Recall that this is the *non-graphical* part of this step of the process.

Summary Statistics ([Tutorial](https://pandas.pydata.org/docs/getting_started/intro_tutorials/06_calculate_statistics.html))
*    Can quickly find various statistics on the whole dataframe using the `describe` **method**.
*    Can look into specific categorical variables using the `groupby` and `agg` **methods**.

```python
df.describe(include='all')
```

```python
# Group by 'FacilityType' (categorical) and aggregate counts for boolean columns
grouped_facility_type = df.groupby('FacilityType').agg(
    facility_count=('FacilityID', 'count'),
    parking_available=('Parking', 'sum'),
    parking_accessible=('ParkingAccessible', 'sum'),
    accessible_facilities=('Accessible', 'sum'),
    avg_latitude=('Latitude', 'mean'),
    avg_longitude=('Longitude', 'mean')
)
print("Grouped by FacilityType:\n", grouped_facility_type)

# Group by 'State' and summarize accessibility and services
grouped_state = df.groupby('State').agg(
    total_facilities=('FacilityID', 'count'),
    accessible_facilities=('Accessible', 'sum'),
    sharps_disposal=('SharpsDisposal', 'sum'),
    drinking_water=('DrinkingWater', 'sum'),
    showers=('Shower', 'sum')
)
print("\nGrouped by State:\n", grouped_state)

# Group by 'PaymentRequired' (boolean) and analyze related features
grouped_payment = df.groupby('PaymentRequired').agg(
    facility_count=('FacilityID', 'count'),
    accessible_facilities=('Accessible', 'sum'),
    adult_change_facilities=('AdultChange', 'sum'),
    baby_care_rooms=('BabyCareRoom', 'sum')
)
print("\nGrouped by PaymentRequired:\n", grouped_payment)

# Group by 'Accessible' and analyze gender-specific and unisex toilet availability
grouped_accessible = df.groupby('Accessible').agg(
    total_facilities=('FacilityID', 'count'),
    male_toilets=('Male', 'sum'),
    female_toilets=('Female', 'sum'),
    unisex_toilets=('Unisex', 'sum'),
    ambulant_toilets=('Ambulant', 'sum')
)
print("\nGrouped by Accessible:\n", grouped_accessible)

# Group by 'ParkingAccessible' to examine related facilities and services
grouped_parking_accessible = df.groupby('ParkingAccessible').agg(
    facility_count=('FacilityID', 'count'),
    accessible_facilities=('Accessible', 'sum'),
    baby_change_facilities=('BabyChange', 'sum'),
    sanitary_disposal=('SanitaryDisposal', 'sum')
)
print("\nGrouped by ParkingAccessible:\n", grouped_parking_accessible)
```

# Data Visualization

Univariate Analysis ([Plotting Tutorial](https://www.kaggle.com/residentmario/univariate-plotting-with-pandas))
*    Plot the distribution of numerical variables with [histograms](https://seaborn.pydata.org/tutorial/distributions.html#plotting-univariate-histograms) or [density plots](https://seaborn.pydata.org/tutorial/distributions.html#kernel-density-estimation).
*    Look for outliers of numerical variables using [violin plots](https://seaborn.pydata.org/generated/seaborn.violinplot.html) or [box plots](https://seaborn.pydata.org/generated/seaborn.boxplot.html).
*    Understand the numerical distribution of categorical variables by calling the `value_counts` **method** on a single Series column.
*    Plot the distribution of categorical variables with [bar charts](https://seaborn.pydata.org/generated/seaborn.barplot.html#seaborn.barplot) (no pie charts please, we hate pie charts).


```python
facility_type_counts = df['FacilityType'].value_counts()
print(facility_type_counts)

# Plotting the distribution of 'FacilityType'
plt.figure(figsize=(10, 6))
sns.countplot(x='FacilityType', data=df)
plt.title('Distribution of Facility Types')
plt.xlabel('Facility Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.show()
```

The distribution of facility types shows that the most common facility type is 'Park or reserve', and the least common being 'Train Station'.

```python
state_counts = df['State'].value_counts()
print(state_counts)

# Plotting the distribution of 'State'
plt.figure(figsize=(12, 6))
sns.countplot(x='State', data=df)
plt.title('Distribution of Public Toilets Across States')
plt.xlabel('State')
plt.ylabel('Number of Toilets')
plt.xticks(rotation=45, ha='right')
plt.show()
```

The Distribution of Public Toilets Across States shows that NSW has the most number public toilets, while ACT has the least number of public toilets.

```python
# Plotting the relationship between 'Accessible' and 'FacilityType'
plt.figure(figsize=(12, 6))
sns.countplot(x='FacilityType', hue='Accessible', data=df)
plt.title('Accessibility of Different Facility Types')
plt.xlabel('Facility Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

The Distribution of Accessibility of Different Facility Types shows that got 'Camping ground', 'Service Station', 'Caravan park', and 'Other' all have a greater non-accessible ratio, while the rest have a greater accessible ratio.

```python
payment_required_counts = df['PaymentRequired'].value_counts()
print(payment_required_counts)

# Plotting the distribution of 'PaymentRequired'
plt.figure(figsize=(6, 4))
sns.countplot(x='PaymentRequired', data=df)
plt.title('Distribution of Payment Required for Toilets')
plt.xlabel('Payment Required')
plt.ylabel('Count')
plt.show()
```

The Distribution of Payment Required for Toilets pleasantly indicates that the overwhelming majority of publiuc toilets do not require payment. :)

```python
# Plotting the relationship between 'Accessible' and 'ParkingAccessible'
plt.figure(figsize=(6, 4))
sns.countplot(x='Accessible', hue='ParkingAccessible', data=df)
plt.title('Accessibility vs. Parking Accessibility')
plt.xlabel('Accessible')
plt.ylabel('Count')
plt.show()
```

The Accessibility vs. Parking Accessibility Visual indicates that for both accessible and non-accessible public toilets, parking is usually inaccessible.

Bivariate Analysis ([Plot Ideas](https://www.kaggle.com/residentmario/bivariate-plotting-with-pandas))
*   Understand the relationships between variables with...
    * [Scatter Plots](https://seaborn.pydata.org/tutorial/relational.html#relating-variables-with-scatter-plots)
    * [Line Plots](https://seaborn.pydata.org/tutorial/relational.html#emphasizing-continuity-with-line-plots)
    * [Regression Plots](https://seaborn.pydata.org/tutorial/regression.html)
    * [Heat Maps](https://seaborn.pydata.org/tutorial/distributions.html#visualizing-bivariate-distributions)
    * and [MORE](https://www.data-to-viz.com/#explore)....



```python
# Calculate proportions
proportions = df.groupby('FacilityType')['Accessible'].value_counts(normalize=True).unstack()

# Analyze proportions of boolean features like Accessible by FacilityType.
proportions.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Proportion of Accessible Facilities by FacilityType')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.legend(title='Accessible', labels=['No', 'Yes'])
plt.show()
```

The Proportion of Accessible Facilities by FacilityType Visual shows that there is a very mixed proportion of accessible vs non-accessible toilets for different facility types. The facility type that is the least accessible is 'Rest area' and the facility type that is the most accessible is 'Swimming pool'.

```python
# Explore relationships between two boolean variables, e.g., ParkingAccessible and Accessible.
# Co-occurrence matrix
boolean_matrix = pd.crosstab(df['ParkingAccessible'], df['Accessible'])
# Heatmap
sns.heatmap(boolean_matrix, annot=True, fmt='d', cmap='coolwarm')
plt.title('Co-occurrence of ParkingAccessible and Accessible')
plt.xlabel('Accessible')
plt.ylabel('ParkingAccessible')
plt.show()
```

The Co-occurence of ParkingAccessible and Accessible Heatmaps shows the majority of inaccessible toilets have inaccessible parking. It also shows that the majority of accessible toilets also have inaccessible parking.
